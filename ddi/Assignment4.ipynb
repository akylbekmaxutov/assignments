```python
import pandas as pd
import numpy as np
from nltk.corpus import wordnet as wn
```

```python
import nltk
nltk.download('wordnet')
```

```python
from gensim.models import Word2Vec
```

```python
simlex = pd.read_csv("SimLex-999.txt", sep="\t")
```

```python
simlex
```

```python
simlex['POS'].unique()
```

```python
word_similarities = np.zeros(len(simlex))
missing = 0
c = 0
for i in zip(simlex['word1'], simlex['word2'], simlex['POS']):
  if len(wn.synsets(i[0])) == 0 or len(wn.synsets(i[1])) == 0:
    missing += 1
  w1 = wn.synset(i[0] + '.' + i[-1].lower() + '.01')
  w2 = wn.synset(i[1] + '.' + i[-1].lower() + '.01')
  word_similarities[c] = wn.path_similarity(w1, w2)
  c += 1
```

```python
missing
```

```python
word_similarities
```

```python
wup_similarities = np.zeros(len(simlex))
c = 0
for i in zip(simlex['word1'], simlex['word2'], simlex['POS']):
  if len(wn.synsets(i[0])) == 0 or len(wn.synsets(i[1])) == 0:
    missing += 1
  w1 = wn.synset(i[0] + '.' + i[-1].lower() + '.01')
  w2 = wn.synset(i[1] + '.' + i[-1].lower() + '.01')
  wup_similarities[c] = wn.wup_similarity(w1, w2)
  c += 1
```

```python
wup_similarities
```

```python
import gensim.downloader as api
dataset = api.load("text8")
model = Word2Vec(dataset)
```

```python
model.wv["hallway"]
```

```python
from scipy.spatial.distance import cosine
cos_similarities = np.zeros(len(simlex))
not_pairs = []
c = 0
for i in zip(simlex['word1'], simlex['word2']):
  if i[0] not in model.wv or i[1] not in model.wv:
    cos_similarities[c] = 0
    c += 1
    not_pairs.append(i)
  else:
    w1 = model.wv[i[0]]
    w2 = model.wv[i[1]]
    cos_similarities[c] = 1 - cosine(w1, w2)
    c += 1
```

```python
not_pairs
```

```python
cos_similarities
```

```python
word_pairs = list(zip(simlex['word1'], simlex['word2']))
```

```python
ranked_pairs_gold = [pair for _, pair in sorted(zip(simlex['SimLex999'], word_pairs))]
ranked_pairs_path = [pair for _, pair in sorted(zip(word_similarities, word_pairs))]
ranked_pairs_wup = [pair for _, pair in sorted(zip(wup_similarities, word_pairs))]
ranked_pairs_cos = [pair for _, pair in sorted(zip(cos_similarities, word_pairs))]
```

```python
from scipy.stats import kendalltau
```

```python
tau_path, p_value_path = kendalltau(ranked_pairs_gold, ranked_pairs_path)
tau_wup, p_value_wup = kendalltau(ranked_pairs_gold, ranked_pairs_wup)
tau_cos, p_value_cos = kendalltau(ranked_pairs_gold, ranked_pairs_cos)
```
