```python
import pandas as pd
import numpy as np
import random
import itertools
```

```python
movies = pd.read_excel('movies.xlsx')
```

```python
movies
```

```python
nums = list(movies["Rank"])
random.shuffle(nums)
```

```python
def get_random_pairs(numbers):
  pairs = list(itertools.combinations(numbers, 2))
  random.shuffle(pairs)
  return pairs
```

```python
pair = get_random_pairs(nums)
```

```python
m1 = []
m2 = []

for p in pair:
  m1.append(movies['Movie'][p[0]-1])
  m2.append(movies['Movie'][p[1]-1])
```

```python
movies_df = pd.DataFrame()
movies_df['left'] = m1
movies_df['right'] = m2
```

```python
movies_df
```

```python
pref = [2, 1, 2, 2, 2, 1, 1, 2, 1, 2, 1, 1, 1, 2, 2, 2, 2, 1, 2, 2, 2, 2, 1,
        2, 2, 1, 2, 1, 2, 2, 2, 1, 2, 1, 2, 1, 2, 2, 1, 2, 1, 2, 2, 1, 2]
```

```python
rank = []
for i in range(len(pref)):
  if pref[i] == 2:
    rank.append(movies['Movie'][pair[i][1]-1])
  else:
    rank.append(movies['Movie'][pair[i][0]-1])
```

```python
movies_df['label'] = rank
```

```python
movies_df
```

```python
model = BradleyTerry(n_iter=100).fit_predict(movies_df)
```

```python
def bootstrap(data):
  df = data.sample(n = len(data), replace=True)
  bt = BradleyTerry(n_iter=100).fit_predict(df)
  return bt
```

```python
for i in range(1000):
  b = bootstrap(movies_df)
  for index in b.index:
    movie_lists[index][i] = b[index]
```

```python
movie_lists
```

```python
final_df = pd.DataFrame()
```

```python
df_name = []
df_25 = []
df_975 = []
```

```python
for key in movie_lists.keys():
  df_name.append(key)
  df_25.append(np.percentile(movie_lists[key], 2.5))
  df_975.append(np.percentile(movie_lists[key], 97.5))
```

```python
final_df['Movie'] = df_name
final_df['2.5 percentile'] = df_25
final_df['97.5 percentile'] = df_975
```
